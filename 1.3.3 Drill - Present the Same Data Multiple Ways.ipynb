```python
# Import the necessary modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Create a dataframe from the csv file, downloaded from Kaggle
df = pd.read_csv('C:\\Users\\maken\\movie_metadata.csv')
```

```python
# Remove NaN values and print the data types
df1 = df.dropna()
print(df.dtypes)
```

Choose one variable and plot that variable four different ways.

```python
# Line plot first
plt.plot(df1['budget'])
plt.show()
```

```python
# Pie chart second
plt.pie(df1['budget'])
plt.show()
```

```python
# Histogram third
n, bin, patches = plt.hist(df1['budget'], bins = 5)
plt.xticks(np.arange(1,30))
plt.show()
```

```python
# Boxplot fourth
plt.boxplot(df1['budget'])
plt.show()
```

Choose two continuous variables, and plot them three different ways.

```python
# Let's use budget and gross
# Scatter plot first
plt.scatter(x=df1['budget'], y=df1['gross'])
plt.show()
```

```python
# Two histograms second
plt.hist(df1['budget'], alpha = .5, label = 'budget')
plt.hist(df1['gross'], alpha = .5, label = 'gross')
plt.legend(loc='upper right')
plt.show()
```

```python
# Two line plots third
plt.plot(df1['budget'], alpha = .5, label = 'budget')
plt.plot(df1['gross'], alpha = .5, label = 'gross')
plt.legend(loc = 'upper right')
plt.show()
```

Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
# Let's use budget and country
# Make a strip plot
sns.set(style = 'whitegrid', color_codes = True)
myplot = sns.stripplot(x=df1['country'], y = df1['budget'])
for item in myplot.get_xticklabels():
    item.set_rotation(60)

labels = df1['country']
plt.show()
```

```python
# Make a box plot second
sns.boxplot(x = df1['country'], y = df1['budget'])
plt.show()
```

```python
# Make a violin plot third
sns.violinplot(x= df1['country'], y = df1['budget'])
plt.show()
```
