```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
import time

# Replace the path with the correct path for your data.
y2015_original = pd.read_csv(
    'C:\\Users\\maken\\Downloads\\LoanStats3d.csv',
    skipinitialspace=True,
    header=1
)
y2015_original.head()
```

```python
y2015 = y2015_original
```

```python
y2015.head()
```

```python
# Drop other columns with many unique variables
#y2015_original.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
 #           'sub_grade', 'addr_state', 'desc'], 1, inplace=True)
```

```python
y2015 = y2015.loc[:,['member_id','loan_amnt','funded_amnt','int_rate','annual_inc','loan_status']]
```

```python
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')
```

```python
y2015['member_id'] = pd.to_numeric(y2015['member_id'], errors = 'coerce')
```

```python
y2015['loan_amnt'] = pd.to_numeric(y2015['loan_amnt'], errors = 'coerce')
```

```python
y2015['funded_amnt'] = pd.to_numeric(y2015['funded_amnt'], errors = 'coerce')
```

```python
y2015['annual_inc'] = pd.to_numeric(y2015['annual_inc'], errors = 'coerce')
```

```python
# Remove two summary rows at the end that don't actually contain data.
y2015 = y2015[:-2]
```

```python
pd.get_dummies(y2015)
```

```python
# Find the unique loan statuses
y2015.loan_status.unique()
```

```python
# Need to change each of the seven unique loan_status values to an id of 1-7
# Use pd.Categorical to assign int ids to the unique loan_status values
y2015['loan_status'] = pd.Categorical(y2015.loan_status)
```

```python
print(y2015.loan_status.cat.codes.unique())
```

```python
y2015.head()
```

```python
y2015.drop(['member_id'], axis=1, inplace=True)
```

```python
y2015 = y2015.dropna()

no_loan_status = y2015.drop(['loan_status'], axis=1)
print(no_loan_status)
```

```python
start_time = time.time()

# This is the model we'll be using.
from sklearn import tree

# A convenience for displaying visualizations.
from IPython.display import Image

# Packages for rendering our tree.
import pydotplus
import graphviz

# Initialize and train our tree.
decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_features=1,
    max_depth=4,
    random_state = 1337
)
decision_tree.fit(no_loan_status, y2015.loan_status.cat.codes)

# Render our tree.
dot_data = tree.export_graphviz(
    decision_tree, out_file=None,
    feature_names=y2015.columns,
    class_names= y2015.loan_status.cat.categories,
    filled=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())



print("--- %s seconds ---" % (time.time() - start_time))
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

start_time = time.time()

rfc = ensemble.RandomForestClassifier()
X = no_loan_status
Y = y2015['loan_status']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

cross_val_score(rfc, X, Y, cv=10)

print("--- %s seconds ---" % (time.time() - start_time))
```
